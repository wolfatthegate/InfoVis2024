```python
import pandas as pd
import json
```

```python
filename = 'a1-mutualfunds.csv'
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(filename)
df.drop('rating', axis = 1, inplace = True)
df.drop('5YR', axis = 1, inplace = True)
df.drop('10YR', axis = 1, inplace = True)
df.drop('Yield', axis = 1, inplace = True)
df.drop('Expense ratio', axis = 1, inplace = True)
df.drop('Mgr tenure', axis = 1, inplace = True)
df.drop('assets', axis = 1, inplace = True)
df.fillna(0, inplace=True)

# Remove commas from the strings and convert them to integers
# df['assets'] = pd.to_numeric(df['assets'].str.replace(',', ''), errors='coerce')

# Rename the column 'old_column_name' to 'new_column_name'
df.rename(columns={'YTD': 'value'}, inplace=True)
df.rename(columns={'3YR': 'threeYR'}, inplace=True)

# Group the DataFrame by the 'Category' column
grouped = df.groupby('Category')

# Convert each group to JSON format and store in a dictionary
json_data = {}
json_data['name'] = 'MUTUAL FUND'
json_data['children'] = []

to_keep = ['name', 'symbol', '3MO', '1YR', 'threeYR', 'value']

for category, group in grouped:
    child_json_data = {}
    child_json_data['name'] = category
    child_json_data['children'] = group[to_keep].to_dict(orient='records')
    json_data['children'].append(child_json_data)

# Write the JSON data to a file
with open('mutualfund.json', 'w') as f:
    json.dump(json_data, f, indent=4)  # Use json.dump for pretty printing

# Alternatively, if you prefer compact JSON output without indentation:
# with open('output.json', 'w') as f:
#     json.dump(json_data, f)
```
